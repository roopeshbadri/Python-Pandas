```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\Badri\Downloads\Ice Cream Ratings.csv")
df=df.set_index('Date')
df
```

```python
print(plt.style.available)
plt.style.use('classic')
```

```python
df.plot(kind='line',title='Ice Cream Ratings', xlabel='Date',ylabel='Scores')
```

```python
df.plot(kind='barh',stacked=True)
```

```python
df.plot.scatter(x='Texture Rating',y='Overall Rating')
```

```python
df.plot(kind='hist',bins=10)
```

```python
df.boxplot()
```

```python
df.plot.area()
```

```python
df.plot(kind='pie',y='Flavor Rating',figsize=(10,6))
```
